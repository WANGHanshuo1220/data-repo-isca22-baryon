```python
''' Required packages. Use python3 -m pip install xxx for all the packages. '''
from pprint import pprint
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm
from matplotlib import pyplot as plt
import easypyplot
%load_ext autoreload
%autoreload 2
```

```python
''' [TODO] Figure properties. Specify the csv file and corresponding labels. '''
workload_list = [
'602.gcc_s',
'605.mcf_s',
'619.lbm_s',
'620.omnetpp_s',
'649.fotonik3d_s',
'654.roms_s',
'657.xz_s'
]
scheme_list = ['Read Hit', 'Write Hit (no overflow)', 'Read Miss', 'Write Miss', 'Write Overflow']
workload_dict = {item: item for item in workload_list}
scheme_dict = {item: item for item in scheme_list}

fig_title = 'Access Breakdown'
fig_name = 'fig_access_breakdown'
fig_ylim = (0, 1)
fig_yticks = list(np.arange(fig_ylim[0], fig_ylim[1]+0.1, 0.25))
# hit_colors = easypyplot.color.color_scale(easypyplot.color.COLOR_SET[1], 2, low=0.65, high=0.8)[::-1]
hit_colors = ['#c8e6c9', '#A2BAA3']
# miss_colors = easypyplot.color.color_scale(easypyplot.color.COLOR_SET[2], 2, low=0.65, high=0.8)[::-1]
miss_colors = ['#e6aecf', '#C494B0']
overflow_color = [easypyplot.color.COLOR_SET[5]]
fig_bar_colors = hit_colors + miss_colors + overflow_color
assert len(fig_bar_colors) == len(scheme_dict)

# pprint(fig_yticks)
```

```python
''' Plot parameters. Change only if you want to resize the graph, etc. '''
fig_bar_width = 0.5
fig_dims = (5, 5)
fig_label_fontsize = 10
ft = fm.FontProperties('Times New Roman')
ft.set_size(10.0)
```

```python
''' Load data from file. '''
csv_path = './layout_access_breakdown/stability_stagecommit.csv'
csv_hd = pd.read_csv(csv_path, sep=',')
fig_xlabels = []
fig_stage_xticks = []
fig_data_stage_2darr = []
xtick = 1
schemedict1 = ['Stage_' + item for item in scheme_dict]
for idx, row in csv_hd.iterrows():
    workload_fullname = row[0]
    if any(workload in workload_fullname for workload in workload_list):
        row_value = []
        for col in row.keys():
            if any(scheme in col for scheme in schemedict1):
                row_value.append(row[col])
        fig_data_stage_2darr.append(row_value)
        fig_xlabels.append(workload_fullname)
        xtick += 2
        fig_stage_xticks.append(xtick)
assert len(fig_data_stage_2darr) == len(fig_xlabels) == len(fig_stage_xticks)
```

```python
''' Load data from file. '''
csv_path = './layout_access_breakdown/stability_stagecommit.csv'
csv_hd = pd.read_csv(csv_path, sep=',')
fig_commit_xticks = []
fig_data_commit_2darr = []
xtick = 1.75
schemedict1 = ['Commit_' + item for item in scheme_dict]
for idx, row in csv_hd.iterrows():
    workload_fullname = row[0]
    if any(workload in workload_fullname for workload in workload_list):
        row_value = []
        for col in row.keys():
            if any(scheme in col for scheme in schemedict1):
                row_value.append(row[col])
        fig_data_commit_2darr.append(row_value)
        xtick += 2
        fig_commit_xticks.append(xtick)
assert len(fig_data_commit_2darr) == len(fig_xlabels) == len(fig_commit_xticks)
```

```python
''' Stage Area Size File loading... '''
csv_path = './layout_access_breakdown/stageareasize.csv'
csv_hd = pd.read_csv(csv_path, sep=',')

fig_16M_xticks = []
fig_data_16M_2darr = []
xtick = 1
schemedict1 = ['16M_' + item for item in scheme_dict]
for idx, row in csv_hd.iterrows():
    workload_fullname = row[0]
    if any(workload in workload_fullname for workload in workload_list):
        row_value = []
        for col in row.keys():
            if any(scheme in col for scheme in schemedict1):
                row_value.append(row[col])
        fig_data_16M_2darr.append(row_value)
        xtick += 2
        fig_16M_xticks.append(xtick)
assert len(fig_data_16M_2darr) == len(fig_xlabels) == len(fig_16M_xticks)

fig_32M_xticks = []
fig_data_32M_2darr = []
xtick = 1.4
schemedict1 = ['32M_' + item for item in scheme_dict]
for idx, row in csv_hd.iterrows():
    workload_fullname = row[0]
    if any(workload in workload_fullname for workload in workload_list):
        row_value = []
        for col in row.keys():
            if any(scheme in col for scheme in schemedict1):
                row_value.append(row[col])
        fig_data_32M_2darr.append(row_value)
        xtick += 2
        fig_32M_xticks.append(xtick)
assert len(fig_data_32M_2darr) == len(fig_xlabels) == len(fig_32M_xticks)

fig_64M_xticks = []
fig_data_64M_2darr = []
xtick = 1.8
schemedict1 = ['64M_' + item for item in scheme_dict]
for idx, row in csv_hd.iterrows():
    workload_fullname = row[0]
    if any(workload in workload_fullname for workload in workload_list):
        row_value = []
        for col in row.keys():
            if any(scheme in col for scheme in schemedict1):
                row_value.append(row[col])
        fig_data_64M_2darr.append(row_value)
        xtick += 2
        fig_64M_xticks.append(xtick)
assert len(fig_data_64M_2darr) == len(fig_xlabels) == len(fig_64M_xticks)

fig_128M_xticks = []
fig_data_128M_2darr = []
xtick = 2.2
schemedict1 = ['128M_' + item for item in scheme_dict]
for idx, row in csv_hd.iterrows():
    workload_fullname = row[0]
    if any(workload in workload_fullname for workload in workload_list):
        row_value = []
        for col in row.keys():
            if any(scheme in col for scheme in schemedict1):
                row_value.append(row[col])
        fig_data_128M_2darr.append(row_value)
        xtick += 2
        fig_128M_xticks.append(xtick)
assert len(fig_data_128M_2darr) == len(fig_xlabels) == len(fig_128M_xticks)
```

```python
''' Stage/Commit Plot '''
pp, fig = easypyplot.pdf.plot_setup(fig_name, fig_dims)
sc_ax = fig.add_subplot(2, 1, 1)
easypyplot.format.turn_off_box(sc_ax)
fig_xlabels = workload_list
h = easypyplot.barchart.draw(
    sc_ax, fig_data_stage_2darr,
    width=fig_bar_width,
    breakdown=True,
    group_names=fig_xlabels,
    xticks=fig_stage_xticks,
    # xticklabelfontsize=fig_label_fontsize,
    xticklabelrotation=40, # Uncomment this for rotated x tick labels
    prop=ft, # Uncomment if easypyplot supports font
    colors=fig_bar_colors)
h2 = easypyplot.barchart.draw(
    sc_ax, fig_data_commit_2darr,
    width=fig_bar_width,
    breakdown=True,
    group_names=fig_xlabels,
    xticks=fig_commit_xticks,
    # xticklabelfontsize=fig_label_fontsize,
    xticklabelrotation=40, # Uncomment this for rotated x tick labels
    prop=ft, # Uncomment if easypyplot supports font
    colors=fig_bar_colors)

''' x axis '''
sc_ax.xaxis.set_tick_params(pad=-5)
sc_ax.set_xlim([sc_ax.get_xticks()[0] - 2, sc_ax.get_xticks()[-1] + 1])
fig_xticks_fix = [item +0.7 for item in fig_stage_xticks]
# fig_xticks_fix[0] -= 0.2
sc_ax.set_xticks(fig_xticks_fix)
for i in range(len(fig_xticks_fix)):
    sc_ax.text(fig_xticks_fix[i] - 0.75, -0.125, 'S    C')
# sc_ax.set_xticklabels(['S', 'C'])
print(sc_ax.get_xticklabels())
sc_ax.set_xticklabels(fig_xlabels, fontsize=fig_label_fontsize, fontproperties=ft, ha='right')
sc_ax.xaxis.set_ticks_position('none') # Comment this and the next line for visible ticks
# sc_ax.tick_params(direction='in') # [in, out, inout]
# sc_ax.set_xlabel(fig_title)
sc_ax.tick_params(axis='both', which='major', pad=5)

''' y axis '''
sc_ax.yaxis.grid(True)
sc_ax.set_ylim(fig_ylim)
sc_ax.set_yticks(fig_yticks)
str_yticks = ['%.2f' % item for item in fig_yticks]
sc_ax.set_yticklabels(str_yticks, fontsize=fig_label_fontsize, fontproperties=ft)
sc_ax.set_ylabel(fig_title, multialignment='center', fontsize=fig_label_fontsize, fontproperties=ft)
sc_ax.set_xlabel('(a) Stage/Commit Phase', fontproperties=ft)

''' legend '''
legend = [scheme_dict[item] for item in scheme_dict]
fig.legend(h, legend, loc='upper center', bbox_to_anchor=(0.5, 1.1),
        ncol=3, frameon=False, prop=ft)


''' Stage Area Size Plot '''
size_ax = fig.add_subplot(2, 1, 2)
easypyplot.format.turn_off_box(size_ax)
fig_xlabels = workload_list
size_bar_width = 0.4
h0 = easypyplot.barchart.draw(
    size_ax, fig_data_16M_2darr,
    width=size_bar_width,
    breakdown=True,
    group_names=fig_xlabels,
    xticks=fig_16M_xticks,
    # xticklabelfontsize=fig_label_fontsize,
    xticklabelrotation=40, # Uncomment this for rotated x tick labels
    prop=ft, # Uncomment if easypyplot supports font
    colors=fig_bar_colors)
h = easypyplot.barchart.draw(
    size_ax, fig_data_32M_2darr,
    width=size_bar_width,
    breakdown=True,
    group_names=fig_xlabels,
    xticks=fig_32M_xticks,
    # xticklabelfontsize=fig_label_fontsize,
    xticklabelrotation=40, # Uncomment this for rotated x tick labels
    prop=ft, # Uncomment if easypyplot supports font
    colors=fig_bar_colors)
h2 = easypyplot.barchart.draw(
    size_ax, fig_data_64M_2darr,
    width=size_bar_width,
    breakdown=True,
    group_names=fig_xlabels,
    xticks=fig_64M_xticks,
    # xticklabelfontsize=fig_label_fontsize,
    xticklabelrotation=40, # Uncomment this for rotated x tick labels
    prop=ft, # Uncomment if easypyplot supports font
    colors=fig_bar_colors)
h3 = easypyplot.barchart.draw(
    size_ax, fig_data_128M_2darr,
    width=size_bar_width,
    breakdown=True,
    group_names=fig_xlabels,
    xticks=fig_128M_xticks,
    # xticklabelfontsize=fig_label_fontsize,
    xticklabelrotation=40, # Uncomment this for rotated x tick labels
    prop=ft, # Uncomment if easypyplot supports font
    colors=fig_bar_colors)
''' x axis '''
# size_ax.xaxis.set_tick_params(pad=-5)
size_ax.set_xlim([size_ax.get_xticks()[0] - 2, size_ax.get_xticks()[-1] + 1])
fig_xticks_fix = [item +0.7 for item in fig_stage_xticks]
# fig_xticks_fix[0] -= 0.2
size_ax.set_xticks(fig_xticks_fix)
size_ax.set_xticklabels(fig_xlabels, fontsize=fig_label_fontsize, fontproperties=ft, ha='right')
size_ax.xaxis.set_ticks_position('none') # Comment this and the next line for visible ticks
# size_ax.tick_params(direction='in') # [in, out, inout]
# size_ax.set_xlabel(fig_title)
# for i in range(len(fig_xticks_fix)):
    # size_ax.text(fig_xticks_fix[i] - 1, -0.125, '32 64 128')
size_ax.tick_params(axis='both', which='major', pad=3)
''' y axis '''
size_ax.yaxis.grid(True)
size_ax.set_ylim(fig_ylim)
size_ax.set_yticks(fig_yticks)
str_yticks = ['%.2f' % item for item in fig_yticks]
size_ax.set_yticklabels(str_yticks, fontsize=fig_label_fontsize, fontproperties=ft)
size_ax.set_ylabel(fig_title, multialignment='center', fontsize=fig_label_fontsize, fontproperties=ft)
size_ax.set_xlabel('(b) Multiple Stage Area Sizes', fontproperties=ft)


''' last resize '''
plt.tight_layout()
easypyplot.format.resize_ax_box(sc_ax, hratio=0.8)
easypyplot.format.resize_ax_box(size_ax, hratio=0.8)
```
